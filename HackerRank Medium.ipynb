```python
def gcd(first, second):
    divisor = first if first < second else second
    while (first % divisor != 0) or (second % divisor != 0):
        step = first % divisor if first % divisor > second % divisor else second % divisor
        divisor -= step
    return divisor

num_lines = int(input())
results = []
for _ in range(num_lines):
    numbers = list(map(int, input().split()))
    result = 0
    for i in range(1, numbers[0]):
        for j in range(i+1, numbers[0] + 1):
            result += gcd(numbers[i], numbers[j])
    results.append(result)

for result in results:
    print(result)
```

```python
def coprime(first, second):
    divisor = second
    while (first % divisor != 0) or (second % divisor != 0):
        step = first % divisor if first % divisor > second % divisor else second % divisor
        divisor -= step
    return divisor == 1

times = int(input())
results = [] 
for i in range(times):
    number = int(input())
    for j in range(number//2, 0, -1):
        if coprime(number, j):
            results.append(j)
            break
for result in results:
    print(result)
```

```python
board_size = 0
board = []

def step(i, j):
    global board_size
    global board
    points = board[i][j][0]
    if board[i][j][1]:
        return points
    rp = step(i+1, j) if i < board_size else 0 
    dp = step(i, j+1) if j < board_size else 0
    points += (rp if rp > dp else dp)
    board[i][j][0] = points
    board[i][j][1] = True
    return points
    

board_size += int(input()) - 1
for _ in range(board_size + 1):
    board.append([[i, False] for i in list(map(int, input().split()))])
print(step(0, 0))
```

```python
def substring(text, pos, sub_size, length):
    count = 0
    if pos + sub_size != length: count += substring(text, pos + 1, sub_size, length)
    if text[pos] == text[pos + sub_size]: count += 1
    return count

def sub_size_change(text, sub_size, length):
    count = 0
    count += substring(text, 0, sub_size, length)
    if sub_size < length: count += sub_size_change(text, sub_size + 1, length)
    return count

text = input()    
print(sub_size_change(text,0, len(text) - 1))
```

```python
def buble(numbers, pos, length):
    if numbers[pos] > numbers[pos + 1] : numbers[pos], numbers[pos + 1] = numbers[pos + 1], numbers[pos]
    if pos + 1 < length : buble(numbers, pos + 1, length)
    
def star_pos_change(numbers, length):
    buble(numbers, 0, length)
    if length > 1: star_pos_change(numbers, length - 1)
        
def create_array(data, numbers, pos, length):
    numbers.append(int(data[pos]))
    if pos < length: create_array(data, numbers, pos+1, length)
    
def print_array(numbers, pos, length):
    print(numbers[pos], end = " ")
    if pos < length: print_array(numbers, pos+1, length)

data = input().split()
numbers = []
length = len(data) - 1
create_array(data, numbers, 0, length)
star_pos_change(numbers, length)
print_array(numbers, 0, length)
```

```python
def conv(num_str):
    if num_str in ["-","+"]:
        num_str += "1"
    return int(num_str)
    
def exchange(line):
    if line[0] != "-":
        line = "+" + line
    result = ""
    for i in range(len(line)):
        if line[i] == "-":
            result += "+"
        elif line[i] == "+":
            result += "-"
        else:
            result += line[i]
    return result
    
def get_result(equation):
    num_val = 0
    x_val = 0
    start = 0
    for i in range(len(equation)):
        if i + 1 == len(equation) or equation[i + 1] in ["+", "-"]:
            if equation[i] == "x":
                x_val += conv(equation[start:i])
            else:
                num_val -= int(equation[start:i+1])
            start = i+1
    if x_val != 0:
        print(num_val/x_val)
    else:
        print("NO")
    
    
times = int(input())
for _ in range(times):
    line = input().split("=")
    if line[0][0] == "x":
        line[0] = "+" + line[0]
    get_result(line[0]+exchange(line[1]))
```

```python
import sys

prime_numbers = [2]

def prime(number):
    global prime_numbers
    for prime_num in prime_numbers:
        if not(number%prime_num):
            return
    prime_numbers.append(number)
    
def binSearch(number):
    
    global prime_numbers
    size = len(prime_numbers)
    
    if number < prime_numbers[0] or number > prime_numbers[size - 1]:
        return False
    
    first = 0
    last = size
    pos = first + last // 2
    while first != size-1 or last != 0:
        if prime_numbers[pos] == number:
            return True
        elif prime_numbers[pos] > number:
            last = pos
        else:
            first = pos
        newpos = (first + last) // 2
        if newpos == pos:
            break
        pos = newpos
    return False

times = int(sys.stdin.readline())
numbers = [int(sys.stdin.readline()) for _ in range(times)]

largest = max(numbers)
if largest > 2:
    for i in range(3, largest+1):
        prime(i)

for number in numbers:
    sys.stdout.write("YES\n" if binSearch(number) else "NO\n")
```
